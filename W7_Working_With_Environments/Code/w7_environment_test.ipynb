We have not yet talked about the `xarray` package. We will do so next week. For now just accept that `xarray` is a package similar to pandas. However unlike pandas, `xarray` is designed to work with multi-dimensional geospatial datasets including satellite data and climate model output. 

The code below loads a _netcdf_ file containing data from the [MODIS instruments](https://modis.gsfc.nasa.gov/) aboard NASA's Terra and Aqua satellite. 

If you have xarray installed, you should be able to successfully execute the code below.  

Look at the output below and try to make sense of it. 

```python
ds = xr.load_dataset('../Data/MOD13A3.061_1km_aid0001_latlon_2020_10km.nc')
ds
```

### Making a simple plot. 

Like *pandas*, *xarray* has a plotting capability. Here I am plotting a variable called `_1_km_monthly_EVI` for the first `.isel('time'=0)` that is in the loaded data. 

```python
ds['_1_km_monthly_EVI'].isel(time=0).plot()
```
